```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '../finetune_IDRiD/_metrics_val.csv'
df = pd.read_csv(file_path, header=None, names=['Acc', 'Sensitivity', 'Specificity', 'Precision', 'AUC-ROC', 'AUC-PR', 'F1', 'MCC', 'Loss'])
```

```python
metrics = ['Acc', 'Sensitivity', 'Specificity', 'Precision', 'AUC-ROC', 'AUC-PR', 'F1', 'MCC', 'Loss']
```

```python
for metric in metrics:
    plt.figure(figsize=(8, 5))
    plt.plot(df.index, df[metric], marker='o', label=metric)
    plt.title(f"{metric} per Epoch")
    plt.xlabel('Epochs')
    plt.ylabel(metric)
    plt.grid(True)
    plt.show()
```
